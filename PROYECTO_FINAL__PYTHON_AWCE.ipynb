```python
def juego():
    """Función juego(): Controla el flujo del juego."""
    #Se inicializa el score a 0 y se establece un bucle que continúa hasta que el usuario falle.
    score = 0
    juego_en_curso = True

    while juego_en_curso== True:
        # Selecciona dos elementos aleatorios
        opcion1, opcion2 = random.sample(datos, 2)

        print(f"\n¿Quién tiene más seguidores en Instagram?")
        print(f"A) {opcion1['nombre']} : {opcion1['descripcion']} de {opcion1['pais']}")
        print(f"B) {opcion2['nombre']} : {opcion2['descripcion']} de {opcion2['pais']}")

        # Obtener respuesta del usuario
        respuesta = input("Escribe 'A' o 'B': ").strip().upper()

        # Comprobar la respuesta
        if opcion1 ['seg_instagram']> opcion2 ['seg_instagram']:
          correcta = "A"
        else:
            correcta = "B"
        if respuesta == correcta :
          score = score + 100
          print("¡Correcto!")
          print (f"Tu puntaje es de: {score}")

        else:
          print("¡Incorrecto!")
          print ("Juego terminado")
          print(f"Puntaje final: {score}") # Mensaje final de puntaje
          juego_en_curso = False
          break





# Iniciar el juego
if __name__ == "__main__":
  reiniciar="SI"
  while reiniciar=="SI":
    print("¡Bienvenido a la trivia de los famosos!")
    print("¿Podrás adivinar quien de estos famosos tiene más seguidores en Instagram?")
    print ("¡COMENCEMOS!")
    juego()
    print("\n ¿Desea reiniciar el juego?")
    reiniciar= input("Escribe SI o NO: ").strip().upper()
    if reiniciar=="SI":
      juego()
    else:
      print("Gracias por jugar")
      break
```
